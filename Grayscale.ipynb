## Grayscale Images in OpenCV

pip install opencv-python

```python
import cv2
```

```python
img_array = np.asarray(bytearray(urllib.request.urlopen('https://cdn.britannica.com/77/142177-050-4E8010A9/Albert-Einstein-1947.jpg').read()),dtype=np.uint8)
img = cv2.imdecode(img_array, cv2.IMREAD_GRAYSCALE)
```

```python
type(img)
```

```python
img.dtype
```

```python
img.shape
```

```python
cv2.imshow('Albert Einstein', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
img[200:400,100:500] = 255
```

```python
cv2.imshow('Albert Einstein', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
cv2.imwrite(r'alber-einstein-openCV-modified.jpg',img)
```
